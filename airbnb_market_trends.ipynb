Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np

# Load the TSV file containing review dates
airbnb_last_review = pd.read_csv('data/airbnb_last_review.tsv', sep='\t')

# Convert the 'last_review' column to datetime format for proper date comparisons
airbnb_last_review['last_review'] = pd.to_datetime(airbnb_last_review['last_review'])

# Find the earliest review date in the dataset
earliest_review = airbnb_last_review['last_review'].min()

# Find the most recent review date in the dataset
most_recent_review = airbnb_last_review['last_review'].max()

# Print the earliest and most recent review dates
print("Earliest review:", earliest_review)
print("Most recent review:", most_recent_review)

# Load the Excel file containing room type information
airbnb_room_type = pd.read_excel('data/airbnb_room_type.xlsx')

# Normalize room_type strings to lowercase and count how many are 'private room'
num_private_rooms = (airbnb_room_type['room_type'].str.lower() == 'private room').sum()

# Print the number of listings that are private rooms
print("Number of Listings of Private rooms:", num_private_rooms)

# Load the CSV file containing listing prices
airbnb_price = pd.read_csv('data/airbnb_price.csv')

# Remove the word 'dollars' from the 'price' column and convert to float
airbnb_price['price'] = airbnb_price['price'].str.replace('dollars', '', regex=False).astype(float)

# Calculate the average listing price and round to 2 decimal places
average_listing_price = airbnb_price['price'].mean().round(2)

# Print the average listing price
print("Average listing price:", average_listing_price)

# Combine all summary information into a single-row DataFrame
review_dates = pd.DataFrame([{
    'first_reviewed': earliest_review,        # Earliest review date
    'last_reviewed': most_recent_review,      # Most recent review date
    'nb_private_rooms': num_private_rooms,    # Number of private room listings
    'avg_price': average_listing_price        # Average price of all listings
}])

# Display the resulting summary DataFrame
print(review_dates)
```
